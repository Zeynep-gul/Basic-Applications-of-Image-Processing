```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('/content/drive/MyDrive/New folder/baboon.tiff')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

gray_image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)

plt.subplot(1, 2, 1)
plt.imshow(image)
plt.title('Orijinal Görüntü')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(gray_image, cmap='gray')
plt.title('Grayscale')
plt.axis('off')

plt.show()
```

```python
r, g, b = cv2.split(image)

def plot_histogram(image, title, color=None):
    plt.hist(image.ravel(), bins=256, color=color, alpha=0.5)
    plt.title(title)
    plt.xlim([0, 256])

plt.subplot(3, 3, 1)
plt.imshow(r, cmap='gray')
plt.title('Kırmızı Kanal')
plt.axis('off')

plt.subplot(3, 3, 2)
plt.imshow(g, cmap='gray')
plt.title('Yeşil Kanal')
plt.axis('off')

plt.subplot(3, 3, 3)
plt.imshow(b, cmap='gray')
plt.title('Mavi Kanal')
plt.axis('off')

# Renkli Görüntü Histogramı
plt.subplot(3, 3, 6)
plot_histogram(image, 'RGB Histogram')

# Kırmızı Kanal Histogramı
plt.subplot(3, 3, 7)
plot_histogram(r, 'Kırmızı Kanal Histogram', color='red')

# Yeşil Kanal Histogramı
plt.subplot(3, 3, 8)
plot_histogram(g, 'Yeşil Kanal Histogram', color='green')

# Mavi Kanal Histogramı
plt.subplot(3, 3, 9)
plot_histogram(b, 'Mavi Kanal Histogram', color='blue')

plt.tight_layout()
plt.show()
```

```python
flipped_vertical = cv2.flip(gray_image, 0)
flipped_horizontal = cv2.flip(gray_image, 1)
flipped_both = cv2.flip(flipped_vertical, 1)

plt.subplot(2, 2, 1)
plt.imshow(gray_image, cmap='gray')
plt.title('Grayscale')
plt.axis('off')

plt.subplot(2, 2, 2)
plt.imshow(flipped_vertical, cmap='gray')
plt.title('Flipped Vertically ')
plt.axis('off')

plt.subplot(2, 2, 3)
plt.imshow(flipped_horizontal, cmap='gray')
plt.title('Flipped Horizontal')
plt.axis('off')

plt.subplot(2, 2, 4)
plt.imshow(flipped_both, cmap='gray')
plt.title('Flipped Both')
plt.axis('off')

plt.show()
```
